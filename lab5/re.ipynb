```python
#7
def snake_to_camel(word):
        
        return ''.join(x.capitalize() or '_' for x in word.split('_'))
with open("C:\\pp2\\lab5\\row.txt", "r", encoding="utf-8") as file:
    text = file.read()
print(snake_to_camel(text))
```

```python
#8 
def splt(string):
    return re.split(r'(?=[A-Z])', string)

with open("C:\\pp2\\lab5\\row.txt", "r", encoding="utf-8") as file:
    text = file.read()
print(splt(text))
```
